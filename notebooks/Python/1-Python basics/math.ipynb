```python
x, y = sym.symbols('x, y')

expr4 = 3 * x + 2
expr5 = x ** 2 + y

display(Math(r"\frac{%s}{%s}" % (sym.latex(expr4), sym.latex(expr5))))
```

```python
x, y, z= sym.symbols('x y z')

expr10 = (3 * x + 2)/ (x**2 + y)

eq = sym.Eq(z, expr10)

display(Math(sym.latex(eq)))
```

```python
# f(x)=x3+2x+1

exprs = x ** 3 + 2 * x + 1

eq = sym.Eq(sym.Symbol("f(x)"), exprs)

eq
```
