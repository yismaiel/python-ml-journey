```python
numbers = []

while True:
    x = input("Enter a number: ").strip()
    if not x:
        break
    
    numbers.append(int(x))

if len(numbers) > 0:
    total = sum(numbers)
    mean = total / len(numbers)
    print(f"sum = {total}, mean = {mean}")
else:
    print("No numbers were entered.")
```
